```python
print("Welcome to my computer quiz game!\n")
play = input("Do you want to play the game?\t")
if play.lower() != "yes":
  exit()
else:
  print("\nLet's get started with the quiz\n")
  score = 0

  #question_1
  
  answer = input("What is the full form of CPU?\t")
  if answer.lower() == "central processing unit":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")
  
  #question_2
  
  answer = input("What is the full form of RAM?\t")
  if answer.lower() == "random access memory":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")
  
  #question_3
  
  answer = input("What is the full form of ROM?\t")
  if answer.lower() == "read only memory":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")

  #question_4
  answer = input("What is the full form of CD?\t")
  if answer.lower() == "compact disc":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")
        
  #question_5
  answer = input("What is the full form of WWW?\t")
  if answer.lower() == "world wide web":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")

  #question_6

  answer = input("What is the full form of URL?\t")
  if answer.lower() == "uniform resource locator":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")

  #question_7

  answer = input("What is the full form of FTP?\t")
  if answer.lower() == "file transfer protocol":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")

  #question_8

  answer = input("What is the full form of ALU?\t")
  if answer.lower() == "arithmetic logic unit":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")

  #question_9

  answer = input("What is the full form of LAN?\t")
  if answer.lower() == "local area network":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")

  #question_10

  answer = input("What is the full form of WAN\t")
  if answer.lower() == "wide area network":
    print("\nYour first answer is correct!\n")
    score += 1
  else:
    print("\nYour first answer is incorrect!\n")

  print("\nYour score is " + str(score))
  print("\nYour percentage is " + str((score/10) * 100))
```
